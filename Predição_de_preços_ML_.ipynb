```python
#PADRONIZANDO E ALTERANDO OS TÍTULOS DAS COLUNAS DA BASE DE DADOS

base_casa = base_casa.rename(columns={"Avg. Area Income":'avg_area_income',
                                     'Avg. Area House Age': 'avg_area_house_age',
                                     'Avg. Area Number of Rooms': 'avg_area_number_rooms',
                                     'Avg. Area Number of Bedrooms': 'avg_number_bedrooms',
                                     'Area Population': 'area_population'})

base_casa.head()
```

```python
#REMOÇÃO DA COLUNA DE ENDEREÇOS PARA FAZER A REGRESSÃO LINEAR, DEIXANDO APENAS DADOS NUMÉRICOS NA BASE DE DADOS

base_casa.drop(["Address"], axis = 1, inplace = True)
base_casa.head()
```

```python
#FAZENDO O HISTOGRAMA PARA MOSTRAR A DISTRIBUIÇÃO DOS PREÇOS

sns.histplot(base_casa['Price'], color = 'black')
```

```python
#DEFININDO AS VARIÁVEIS NOS EIXOS 'X' E 'Y' PARA DEFINIR AS FUNÇÕES: NESSE CASO, AS VARIÁVEIS EM 'X' QUE VÃO TENTAR PREVER O PREÇO

X = base_casa[[ 'avg_area_income',
                'avg_area_house_age',
                'avg_area_number_rooms',
                'avg_number_bedrooms',
                'area_population']]

Y = base_casa[['Price']]
```

```python
#IMPORTANDO A BASE DE TREINO E TESTE PARA O MODELO

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
#VERIFICANDO O TAMANHO DAS BASES SELECIONADAS

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#TREINAMENTO DO MODELO

from sklearn.linear_model import LinearRegression

ln = LinearRegression()
```

```python
ln = ln.fit(X_train, Y_train)
```

```python
#TENTATIVA DE PREDIZER O 'X TRAIN'

Y_pred = ln.predict(X_test)
```

```python
#VERIFICAÇÃO A EFICIÊNCIA DO MODELO

from sklearn.metrics import r2_score

r = r2_score(Y_test, Y_pred)

print('r_quadrado:', r)
```
